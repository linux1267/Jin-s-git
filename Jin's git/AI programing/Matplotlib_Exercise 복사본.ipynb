```python
import matplotlib.pyplot as plt
import numpy as np
```

  (1) plt.plot으로 임의의 그래프 1개를 만드세요

```python
x = np.array([1,2,3,4,5])
y = x**2
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Q1")
plt.show()
```

(2) plt.bar으로 임의의 그래프 1개를 만드세요

```python
x2 = np.array([3,5,76,2,23])
plt.bar(x2,x2+5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Q2")
plt.show()
```

  (3) plt.scatter으로 임의의 그래프 1개를 만드세요

```python
x3 = np.random.rand(50)
y3 = np.random.rand(50)

colors = np.random.randint(0,255,50)  #각각의 점들을 다른색( 0~255가지 색 )으로 50개의 점에 색을 지정
sizes = np.pi*1000*np.random.rand(50)  #각각의 점들의 사이즈를 랜덤하게

plt.scatter(x3,y3,colors,sizes)
plt.show()
```

  (4) plt.hist 으로 임의의 그래프 1개를 만드세요

```python
data = np.random.randn(500)
plt.hist(data,bins=20,color='red') #20개의 bins에 500개의 데이터들이 분포되어 있다. 
plt.title('Q4')
plt.show()
```

  (6) plt.imshow() 으로 임의의 그림 1개를 만드세요

```python
img = np.random.randint(0,255,(100,100))  #0~255사이의 값들로 크키가 100by100인 값들
plt.imshow(img)
plt.title('Q6')
plt.colorbar()
```

  (7) 임의의 함수를 정하고 이를 미분한 함수를 같이 만드세요. 이 둘을 한개의 그래프에 같이 표시하고 미분값이 의미하는 바를 적어보세요.

```python
x7 = np.linspace(0,10,50)  #x7의 값을 0부터 10사이의 값들 중 50개를 뽑는다. (0.2 간격으로)
y7_1 = x7**3 + 6*x7 + 3 
y7_2 = 3*x7**2 + 6  #위의 함수를 미분한 형태

plt.plot(x7,y7_1,linestyle = "-", color = "red", label = "$x^3 + 6x + 3$")
plt.plot(x7,y7_2,linestyle = ":", color = "blue", label = "$3x^2 + 6$")
plt.legend(shadow = True, borderpad = 2 )
plt.show() #$를 사용하면 수식을 표현할 수 있다. shadow를 사용하면 라벨의 그림자를 줄 수 있다. 
```
